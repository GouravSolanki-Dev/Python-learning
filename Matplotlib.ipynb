# **`MatPlotLib Tutorial`**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

Create publication quality plots.

Make interactive figures that can zoom, pan, update.

Customize visual style and layout.

Export to many file formats.

Embed in JupyterLab and Graphical User Interfaces.

Use a rich array of third-party packages built on Matplotlib.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# Plot the Graph
plt.scatter(x,y,c='y')
```

```python
plt.scatter(x,y,c='y')
#give some labels

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('scatter.jpg')
```

```python
y=x*x
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.plot(x,y,'ro')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('2D Plot')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'y*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
plt.subplot(2,2,4)
plt.plot(x,y,'go-')
```

# Histogram

```python
a=np.array([1,20,32,12,58,56,85,95,15,21,48,5,48,99,87,98,65])
plt.hist(a)
plt.title('Histogram')
```

# Pie

```python
labels='Python','C++','Ruby','Java'
sizes=[198,123,98,136]
col=['Green','Red','Yellow','Blue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=col,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
```
