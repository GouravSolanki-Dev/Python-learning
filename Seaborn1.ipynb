# **Seaborn**

> Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
 
>**Distributed plots**
>> DistPlot -- univariant

>> JoinPlot -- BiVariant

>> PairPlot -- Multivariant


```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

    Corelation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Univariant Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',color='green')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',color='red')
```

PairPlot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,kind='reg',hue='sex')
```

```python
sns.pairplot(df,hue='sex')
```



---
DistPlot
---


```python
sns.displot(df['tip'])
```

```python
sns.distplot(df['tip'],color='green')
```

## **Categorical features**

**Count plot**

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

Box Plot

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

Violin plot

```python
sns.violinplot(y='total_bill',x='day',data=df,palette='rainbow')
```
